```python
#Grouping and aggregating
df.groupby('AGE')['SOLD PRICE'].mean()
```

```python
sold_price_age=df.groupby('AGE')['SOLD PRICE'].mean().reset_index()
print(sold_price_age)
```

```python
sold_price_by_age_role=df.groupby(['AGE','PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
print(sold_price_by_age_role)
```

### Joining DataFrame

```python
sold_price_comparision=sold_price_by_age_role.merge(sold_price_age,on='AGE',how='outer')
```

```python
(sold_price_comparision)
```

### Renaming the columns

```python
sold_price_comparision.rename(columns={'SOLD PRICE_x':'SOLD_PRICE_AGE_ROLE','SOLD PRICE_y':'SOLD_PRICE_AGE'},inplace=True)
```

```python
sold_price_comparision
```

### APPLYING OPERATION TO MULTIPLE COLUMNS

```python
sold_price_comparision['change']=sold_price_comparision.apply(lambda rec:(rec.SOLD_PRICE_AGE_ROLE-rec.SOLD_PRICE_AGE)/rec.SOLD_PRICE_AGE,axis=1)
```

```python
sold_price_comparision
```
